```python
EX1 = {
    'S': {'A': 1,'B': 2,'C': 1},
    'A':{'C':1,'D':2,'G':6},
    'B': {'D':2,'E':3},
    'C':{'G':8},
    'D':{'E':1},
    'E':{'G':1},
    'G':{}
}
```

```python
# Uniform Cost Search

from queue import PriorityQueue


def ucs(snode, dnode, GRAPH):
    expanded = [] # to store expanded nodes
    frontier = PriorityQueue()

    node = (0,snode,[snode]) #(g(n) = path cost, city name, path)
    frontier.put(node)
    i = 1
    while not frontier.empty():
        node = frontier.get() # select a node for expansion
        u = node[1]
        if u in expanded:
          continue
        print()
        print(u+ " selected. Cost: "+ str(node[0]))
        if u == dnode:# goal test
            print("*** GOAL TEST SATISFIED ***")
            print()
            return node # path
        expanded.append(u)
        print('['+str(i)+']: '+u+" expanding...")
        i = i +1
        for v in GRAPH[u].keys():#expanding
            if v not in expanded:
                g = node[0] + GRAPH[u][v]
                path = node[2] + [v]
                frontier.put((g,v,path))
                print(u + "->"+v + ", Path: "+ str(path) + ", Cost:"+ str(g))
        #print(frontier.queue)
    print('Failed')
```

```python
ucs('S', 'G', EX1) # change source and destination
```

```python
from queue import PriorityQueue

def gbfs(snode, dnode, GRAPH, heur):
    expanded = [] # to store expanded nodes
    frontier = PriorityQueue()

    snode = (heur[snode],snode,[snode],0) # (h, city, path, path cost)
    frontier.put(snode)
    i =1
    while not frontier.empty():
        node = frontier.get()
        u = node[1]
        if u in expanded:
          continue

        print(u+ " selected. Cost: "+ str(node[0]))
        if u== dnode:#Goal Test
            print(node)
            print(expanded)
            return node #path
        expanded.append(u)
        print('['+str(i)+']: '+u+" expanding...")
        i = i+1
        for v in GRAPH[u].keys():
            if v not in expanded:
                g = node[3] + GRAPH[u][v]
                path = node[2]+[v]
                frontier.put((heur[v],v,path, g))
                print(u + "->"+v + ", Path: "+ str(path) + ", h:"+ str(heur[v]))
    print('Failed')
```

```python
gbfs('S', 'G', EX1, h1) # change source only
```

```python
from queue import PriorityQueue

def astar(snode, dnode, GRAPH, heur):

    frontier = PriorityQueue()
    frontier.put((heur[snode],0,snode,[snode])) # (f,g,node,path)

    expanded = []
    i = 1

    while not frontier.empty():

        f,g,u,path = frontier.get()

        if u in expanded:
            continue

        print()
        print(u+" selected. g:"+str(g)+" h:"+str(heur[u])+" f:"+str(f))

        if u == dnode:
            print("*** GOAL TEST SATISFIED ***")
            return (f,g,u,path)

        expanded.append(u)

        print('['+str(i)+']: '+u+" expanding...")
        i += 1

        for v in GRAPH[u].keys():

            if v not in expanded:

                g2 = g + GRAPH[u][v]
                f2 = g2 + heur[v]
                path2 = path + [v]

                frontier.put((f2,g2,v,path2))

                print(u+"->"+v+", Path:"+str(path2)+
                      ", g:"+str(g2)+
                      ", h:"+str(heur[v])+
                      ", f:"+str(f2))
```

```python
astar('S', 'G', EX1, h1)
```
