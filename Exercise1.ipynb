```python
# Table Driven Agent

table = {
    ('A', 'Dirty'): 'Suck',
    ('A', 'Clean'): 'Right',
    ('B', 'Dirty'): 'Suck',
    ('B', 'Clean'): 'Left'
}
```

```python
percepts = []

def lookup(percept, table):
    return table.get(percept)

def table_driven_agent(percept):
    print("Perception:", percept)
    percepts.append(percept)
    action = lookup(percept, table)
    return action
```

```python
# Simple Reflex Agent

def simple_reflex_agent(percept):
    location = percept[0]
    status = percept[1]

    if status == 'Dirty':
        return 'Suck'
    elif location == 'A':
        return 'Right'
    elif location == 'B':
        return 'Left'
```

```python
import random

Location = random.choice(['A', 'B'])
Condition = random.choice(['Clean', 'Dirty'])

while True:
    print("\nCurrent Location:", Location)
    print("Current Condition:", Condition)

    # Choose one agent
    action = simple_reflex_agent((Location, Condition))
    # action = table_driven_agent((Location, Condition))

    print("Action:", action)

    cmd = input("Continue? (yes/no): ")
    if cmd == 'no':
        break

    if action == 'Right':
        Location = 'B'
        Condition = random.choice(['Clean', 'Dirty'])
    elif action == 'Left':
        Location = 'A'
        Condition = random.choice(['Clean', 'Dirty'])
    else:
        Condition = 'Clean'
```
